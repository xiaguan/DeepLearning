```python
import pandas as pd
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
data = pd.read_csv('train.csv')
data.columns
```

### relplot 散点图
x,y,data(DataFrame)为必须
可选:
hue:数据中的名称
rwo,col：数据中变量的名称
col_wrap:int
size:数据中的名称

```python
pre = pd.concat((data.loc[:,'GrLivArea':'GrLivArea'],data.loc[:,'SalePrice':'SalePrice']),axis=1)
sns.relplot(x='GrLivArea',y='SalePrice',data=pre)
```

```python
var = 'SaleType'
pre = pd.concat((data.loc[:,'SaleType':'SaleType'],data.loc[:,'SalePrice':'SalePrice']),axis=1)
sns.relplot(x=var,y='SalePrice',hue='SaleType',data=pre)
```

```python
var = 'SaleType'
pre = pd.concat((data.loc[:,:'SaleType'],data.loc[:,'SalePrice':'SalePrice']),axis=1)
sns.relplot(x='GrLivArea',y='SalePrice',hue=var,data=pre,style=var,size='Id')
```

### scatterplot 散点图
x,y为数据中变量的名称


```python
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
a = np.random.normal(scale=10,size=(1000,1))
a
```

```python
result = np.expm1(a)
result
```

```python
pre = pd.concat((pd.DataFrame(result,columns=['y']),pd.DataFrame(a,columns=['x'])),axis=1)
pre
```

```python
sns.relplot(data=pre,kind='line',x='x',y='y')
```
